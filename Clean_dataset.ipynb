1. airline :

2. flight : Flight number or code.

3. source_city : City from where the flight departs.

4. departure_time : Departure time (can be actual time or bucketed like Morning, Afternoon, Evening).

5. stops : Number of stops (e.g., Non-stop, 1 stop, 2 stops).

6. arrival_time : Arrival time (or bucketed category).

7. destination_city : City where the flight lands.

8. class : Travel class (e.g., Economy, Business).

9. duration : Duration of the flight journey.

10. days_left : Number of days left until the journey date (from booking).

11. price : Ticket price (target variable in many analyses).

**Data** **Analysis**

**Analysing the number of stops**



```python
Info['stops'].unique()
```

```python
count=Info.stops.value_counts()
plt.pie(count,labels=count.index,autopct=None)
plt.title('Number of stops')
plt.show()
```

**Conclusion**:

Basically there are flights that have one stops

**Analysing the Departure and arrival time**

```python
depart=Info.departure_time.value_counts()
plt.pie(depart,labels=depart.index,autopct=None)
plt.title('Departure Time')
plt.show()
```

**Conclusion**:


There are mainly flghts that departs in morning time

```python
arr=Info.arrival_time.value_counts()
plt.pie(arr,labels=arr.index,autopct=None)
plt.title('Arrival Time')
plt.show()
```

**Conclusion**:

The flight's arriving time is evenly distributed

**Analysing the destination and source city**

```python
depart=Info.destination_city.value_counts()
plt.pie(depart,labels=depart.index,autopct=None)
plt.title('Destination City')
plt.show()
depart1=Info.source_city.value_counts()
plt.pie(depart,labels=depart1.index,autopct=None)
plt.title('Source City')
plt.show()
```

**Conclusion**:

There are almost equal flights are cities

**Analysing the number of airlines**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='airline', data=Info)
plt.xlabel('Airlines')
plt.ylabel('Number of Airlines')
plt.show()
```

**Conclusion**:

There are mainly airlines of Vistara and Air Lines and so on

**Analysing the type of class**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='class', data=Info)
plt.xlabel("Class")
plt.ylabel("Number of Flights")
plt.title("Count of Flights per Class")
plt.show()
```

**Conclusion**:

There are flights mainly for economy class

```python
plt.figure(figsize=(8,5))
plt.hist(Info['days_left'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Days Left")
plt.ylabel("Number of Flights")
plt.title("Distribution of Days Left for Flights")
plt.show()
```

**Analysing price with respect to number of flights**

```python
plt.figure(figsize=(8,5))
plt.hist(Info['price'], bins=5, color='lightgreen', edgecolor='black')
plt.xlabel("Price")
plt.ylabel("Number of Flights")
plt.title("Price Distribution of Flights")
plt.show()
```

**Conclusion**:

:There are mainly flight in between 0 to 2000 price range

**Analysing the price according to days left**

```python
sns.lineplot(x='days_left', y='price', data=Info)
plt.xlabel("Days Left")
plt.ylabel("Price")
plt.title("Days Left vs Price")
```

**Conclusion**:

Prices are more for airlines that have less time to depart

**Analysing the numeric data**

```python
sns.lineplot(x='duration',y='price', data=Info)
plt.xlabel("Duration")
plt.ylabel("Cost of flight")
plt.title("Duration vs Cost")
```

**Conclusion**:

Cost of flights is more for long duration of flights

```python
sns.countplot(x='source_city', hue='destination_city', data=Info)
plt.xlabel("Source City")
plt.ylabel("Number of flight")
plt.title("Number of flights vs source city")
```

Conclusion:

There are major flights between mumbai and delhi

```python
correlation = Info.select_dtypes(include=np.number).corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```

Conclusion:

days_left vs price → negative means closer to travel,price increases

duration vs price → positive means longer flights,cost more

***Overall Conclusion***:


*   Major of the flights have one or more stops
*   Mostly flights departure flight is in morning

*   There are major flight between mumbai and Delhi
*   Mainly flight are under 2000 price
